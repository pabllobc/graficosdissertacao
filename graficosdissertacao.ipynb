```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
cenarios = ['TCP/H264/GS', 'UDP/H264/RT', 'UDP/H264/GS']
media = [37.0000, 34.0000, 33.0000]
moda = [33.0000, 33.0000, 33.0000]
desvio_padrao = [25.0197, 4.3647, 9.0881]

# Posição das barras
x = np.arange(len(cenarios))
largura = 0.25

# Criando a figura e o gráfico
fig, ax = plt.subplots(figsize=(10, 6))
barras_media = ax.bar(x - largura, media, largura, label='Média (M)', color='blue')
barras_moda = ax.bar(x, moda, largura, label='Moda (Mo)', color='red')
barras_desvio = ax.bar(x + largura, desvio_padrao, largura, label='Desvio Padrão (DP)', color='yellow')

# Adicionando os valores acima das barras
for barras in [barras_media, barras_moda, barras_desvio]:
    for barra in barras:
        altura = barra.get_height()
        ax.text(barra.get_x() + barra.get_width() / 2, altura + 0.5, f'{altura:.4f}', ha='center', va='bottom', fontsize=8)

# Configurações do gráfico
ax.set_title('Experimento com Wifi - Latência (E2E) ms', fontsize=10)
ax.set_xticks(x)
ax.set_xticklabels(cenarios, rotation=0, fontsize=10)
ax.set_ylabel('Latência (ms)', fontsize=8)
ax.set_xlabel('Cenários 1, 2 e 3', fontsize=8)
ax.legend(fontsize=8)
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Exibir o gráfico
plt.tight_layout()
plt.show()

# Salvar gráfico
n_fig1 = '1_wifi_cenarios-123_E2E.png'
fig.savefig(n_fig1)
```
